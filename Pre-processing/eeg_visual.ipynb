```python
from numpy import genfromtxt
import numpy as np
import matplotlib.pyplot as pp
#from sklearn.preprocessing import normalize

def normalize(v):
    norm=np.linalg.norm(v, ord=1)
    if norm==0:
        norm=np.finfo(v.dtype).eps
    print(norm)
    return v/norm
```
